# Label Encoding

> Label Encoding assigns each unique catagorical value  a numerical code







```python
classes=['ClaasA','ClassB','ClassC','ClassD']
instances=['ClassA','ClassB','ClassC','ClassD','ClassA','ClassC']

#Encoded the Data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
encoded_data=le.fit_transform(instances)
print(f"Encoded Data:{encoded_data}")

#Decoded the Data
decoded_data=le.inverse_transform(encoded_data)
print(f"Decoded Data(Orginal Data):{decoded_data}")


```

# Ordinal Encoding

> image.png

```python
from sklearn.preprocessing import OrdinalEncoder
data=['good','bad','excellent','average','good','excellent','bad']
catagories=[['bad','average','good','excellent']] #catagories will be in data
df2=pd.DataFrame(data,columns=['Reviews']) #reviews
#encoder
encoder=OrdinalEncoder(categories=catagories)
encoder_data=encoder.fit_transform(df2)
encoder
#decoder
encoder.inverse_transform(encoder_data)
```
